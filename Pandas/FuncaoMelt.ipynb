```python
import pandas as pd
```

```python
df = pd.DataFrame({'A':{0:'a', 1:'b', 2:'c'}, 
                      'B':{0:1, 1:3, 2:5},
                      'C':{0:2, 1:4, 2:6}})
df
```

```python
pd.melt(df, id_vars = ['A'], value_vars = ['B', 'C'], var_name = 'Nome da Coluna', value_name = 'Valor da Coluna')
```

```python
data = {'Localização': ['CidadeA', 'CidadeB'],
       'Temperatura': ['Prevista', 'Atual'],
       'Set-2019':[30, 32],
       'Out-2019':[45,43],
       'Nov-2019':[24,22]}

df = pd.DataFrame(data)

df
```

```python
melt_df = pd.melt(df, id_vars= ['Localização', 'Temperatura'], var_name = 'Data', value_name = 'Valor')
melt_df
```
