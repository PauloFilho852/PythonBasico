```python
from openai import OpenAI
import os
from IPython.display import Audio, display
import requests
import dotenv
```

```python
dotenv.load_dotenv()
client = OpenAI()
```

```python
path = 'Files\\pt.mp3'
```

```python
#Atualmente há dois modelos tts-1 (recomendado para tempo real) e tts-1-hd

#Parâmetros básicos:
#model - modelo a ser utilizado
#voice - tibre de voz
#input - texto a ser tranformado

#Parâmetros adicionais:
#speed - Padrão é 1. Velocidade de reprodução do áudio.Varia de 0.25 até 4.
#response_format - Padrão é mp3. Há o opus (conexões de baixa latência), aac, flac (sem compressão)

def audio_create(model = 'tts-1', voice = 'alloy', input = None, speed = 1.0, response_format = 'mp3'):
    
    response = client.audio.speech.create(
        model = model,
        voice = voice,
        input = input,
        speed = speed,
        response_format = response_format
    )
    
    return response    
```

```python
input = 'Vem, vamos viver! Temos muito ainda por fazer. Não olhe pra traz. Apenas começamos! O mundo começa agora...'

#A versão atual do TTS-1 permite até 3 requisições seguidas, a partir de então, deve-se esperar 20 segundos.
audio = audio_create(input = input)
```

```python
audio.write_to_file(path)
```
