```python
from openai import OpenAI
import os
from IPython.display import Audio, display
import dotenv
```

```python
dotenv.load_dotenv()
client = OpenAI()
```

```python
path = 'Files\\camoes.m4a'
```

```python
#TRANSCRIÇÃO DE ÁUDIO EM TEXTO
```

```python
#Parâmetros básicos
#model - Modelo utilizado
#file - Arquivo a ser transcrito.

#Parâmentros adicionais
#language - Linguagem do áudio.
#prompt - Descrição do áudio, para melhor transcrição, incluindo pontuação. Também é usado para indicar quebra de áudio. Tem que estar na linguagem do
#do áudio.
#response_format - Formato da resposta. Padrão é JSON. Existem o json, text, str, vtt e ..json
#temperature - Por padrão é 0. Varia de 0 a 1. Valores maiores aumentam a criatividade ou aleatoriedade.


def transcript(model = 'whisper-1', audio_file = None, language = 'pt', prompt = None, response_format = 'text', temperature = 0):
    
    response = client.audio.transcriptions.create(
        model = model,
        file = audio_file,
        language = language,
        prompt = prompt,
        response_format = response_format,
        temperature = temperature
        )
    
    return response
```

```python
audio_file = open(path, 'rb')

prompt = 'É um poema.'

transcription = transcript(audio_file = audio_file, prompt = prompt)

audio_file.close()
```

```python
display(Audio(path, autoplay = False))
print(transcription)
```

```python
#Não possui o parâmetro language.
#Traduz de qualquer linguagem para o inglês (até a presente versão)

def translation(model = 'whisper-1', audio_file = None, prompt = None, response_format = 'text', temperature = 0):
    
    response = client.audio.translations.create(
        model = model,
        file = audio_file,
        prompt = prompt,
        response_format = response_format,
        temperature = temperature
        )
    
    return response
```

```python
audio_file = open(path, 'rb')

prompt = 'É um poema.' 

translated_audio = translation(audio_file = audio_file, prompt = prompt)

audio_file.close()
```

```python
display(Audio(path, autoplay = False))
print(translated_audio)
```
