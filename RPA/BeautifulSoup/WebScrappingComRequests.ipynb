```python
import requests
from bs4 import BeautifulSoup as BS
```

```python
page = requests.get('https://www.gabrielcasemiro.com.br')

ok = False

if(page.status_code == 200):
   ok = True
else:
    print('HTTP Error: ', page.status_code)

if ok :
     soup = BS(page.text)
```

```python
soup
```

```python
links = soup.find_all('a')
len(links)
```

```python
for a in links:
    #print(a['href']) --> pode quebrar se n√£o encontrar um href pode quebrar
    print(a.get('href', '')) # se n√£o encontrar, retorna um string vazia 
```
