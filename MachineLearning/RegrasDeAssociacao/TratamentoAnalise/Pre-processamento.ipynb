```python
import pandas as pd
import numpy as np
import warnings
import os
```

```python
path = f'{os.getcwd()}\\Dados\\'
warnings.filterwarnings('ignore')

df = pd.read_csv(path + 'Groceries_dataset.csv', sep = ',', encoding = 'iso-8859-1')
df.head()
```

```python
df_2 = df.groupby(['Member_number', 'itemDescription'])['itemDescription'].count()
df_2
```

```python
df_2 = df_2.unstack()
df_2
```

```python
df_2 = df_2.fillna(0).applymap(lambda x: 1 if x > 0 else 0 )
df_2
```

```python
df_2.to_csv(path + 'dados_tratados.csv', sep= ';', index = False)
```
