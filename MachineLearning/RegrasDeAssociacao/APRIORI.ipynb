```python
#pip install mlxtend
from pyECLAT import Example1, ECLAT
from mlxtend.frequent_patterns import apriori, association_rules
import warnings
```

```python
warnings.filterwarnings('ignore')
```

```python
dados = Example1().get()
```

```python
dados
```

```python
#Transformando os dados para efeito de cálculos.
eclat = ECLAT(data = dados, verbose = True)
```

```python
dados_bin = eclat.df_bin
dados_bin
```

```python
eclat.uniq_
```

```python
#Gerando a associação
associacao = apriori(dados_bin, min_support=0.1, use_colnames = True)
```

```python
#Colocando e exibindo em ordem  mais frequentes (support)
associacao.sort_values('support', ascending=False)
```

```python
#São formados conjuntos associativos e são tomadas as taxas de frequências (support) que ocorrem.
#Subconjuntos são tomados destes conjuntos.

#Gerando regras de associação (Ordenação com base na métrica, tomando por base os subconjuntos de cada 
#conjunto associativo)

regras = association_rules(associacao, metric = 'lift')
regras
```

```python
#Ordenando pela somente pela métrica (support)
resultado = regras.sort_values('support', ascending = False)
resultado.tail()
```
