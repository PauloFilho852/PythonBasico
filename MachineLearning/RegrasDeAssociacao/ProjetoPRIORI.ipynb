```python
import pandas as pd
import numpy as np
import warnings
import os
from pyECLAT import Example1, ECLAT
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
path = f'{os.getcwd()}\\Tratamento e Analise\\Dados\\'
warnings.filterwarnings('ignore')

dados = pd.read_csv(path + 'dados_tratados.csv', sep = ';')
dados.head()
```

```python
associacao = apriori(dados, min_support = 0.05, use_colnames=True)
associacao.sort_values('support', ascending = False)
associacao.head()
```

```python
#----Support (Suporte):O suporte varia de 0 a 1.
#O suporte indica a frequência absoluta com que um conjunto de itens (A e B juntos) aparece nas transações.

#----Confidence (Confiança):A confiança varia de 0 a 1.
#A confiança mede a probabilidade de que o consequente (B) ocorra em uma transação, dado que o antecedente (A) também ocorreu.
#Quanto maior a confiança, mais forte é a associação entre A e B.
#Uma confiança de 1 indica que B sempre ocorre quando A ocorre.

#----Lift (Elevação): O lift varia de 0 a infinito.
#O lift compara a frequência observada de A e B ocorrendo juntos com a frequência esperada se a associação fosse independente.
#Um lift maior que 1 indica uma associação positiva entre A e B (eles ocorrem juntos mais frequentemente do que o esperado ao 
#acaso).Um lift igual a 1 indica independência (sem associação).Um lift menor que 1 indica uma associação negativa 
#(A e B ocorrem juntos menos frequentemente do que o esperado ao acaso).

#----Leverage (Alavancagem): A alavancagem pode variar de -1 a 1.
#A alavancagem mede a diferença entre a frequência observada de A e B ocorrendo juntos e a frequência esperada se fossem indepen
#dentes. Um valor positivo de alavancagem indica uma associação positiva entre A e B. Um valor negativo indica uma associação 
#negativa ou ausência de associação.

#----Conviction (Convicção): A convicção pode variar de 0 a infinito.
# A convicção mede o grau de dependência do consequente (B) na presença do antecedente (A). Quanto maior a convicção, mais 
#forte é a dependência de B em relação a A. Uma convicção de 1 indica independência (ausência de associação entre A e B).
```

```python
regras = association_rules(associacao, metric = 'lift')
regras.head()
```
