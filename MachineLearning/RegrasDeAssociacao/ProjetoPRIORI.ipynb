```python
import pandas as pd
import numpy as np
import warnings
import os
from pyECLAT import Example1, ECLAT
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
path = f'{os.getcwd()}\\Tratamento e Analise\\Dados\\'
warnings.filterwarnings('ignore')

dados = pd.read_csv(path + 'dados_tratados.csv', sep = ';')
dados.head()
```

```python
associacao = apriori(dados, min_support = 0.05, use_colnames=True)
associacao.sort_values('support', ascending = False)
associacao.head()
```

```python
#----Support (Suporte):O suporte varia de 0 a 1.
#O suporte indica a frequÃªncia absoluta com que um conjunto de itens (A e B juntos) aparece nas transaÃ§Ãµes.

#----Confidence (ConfianÃ§a):A confianÃ§a varia de 0 a 1.
#A confianÃ§a mede a probabilidade de que o consequente (B) ocorra em uma transaÃ§Ã£o, dado que o antecedente (A) tambÃ©m ocorreu.
#Quanto maior a confianÃ§a, mais forte Ã© a associaÃ§Ã£o entre A e B.
#Uma confianÃ§a de 1 indica que B sempre ocorre quando A ocorre.

#----Lift (ElevaÃ§Ã£o): O lift varia de 0 a infinito.
#O lift compara a frequÃªncia observada de A e B ocorrendo juntos com a frequÃªncia esperada se a associaÃ§Ã£o fosse independente.
#Um lift maior que 1 indica uma associaÃ§Ã£o positiva entre A e B (eles ocorrem juntos mais frequentemente do que o esperado ao 
#acaso).Um lift igual a 1 indica independÃªncia (sem associaÃ§Ã£o).Um lift menor que 1 indica uma associaÃ§Ã£o negativa 
#(A e B ocorrem juntos menos frequentemente do que o esperado ao acaso).

#----Leverage (Alavancagem): A alavancagem pode variar de -1 a 1.
#A alavancagem mede a diferenÃ§a entre a frequÃªncia observada de A e B ocorrendo juntos e a frequÃªncia esperada se fossem indepen
#dentes. Um valor positivo de alavancagem indica uma associaÃ§Ã£o positiva entre A e B. Um valor negativo indica uma associaÃ§Ã£o 
#negativa ou ausÃªncia de associaÃ§Ã£o.

#----Conviction (ConvicÃ§Ã£o): A convicÃ§Ã£o pode variar de 0 a infinito.
# A convicÃ§Ã£o mede o grau de dependÃªncia do consequente (B) na presenÃ§a do antecedente (A). Quanto maior a convicÃ§Ã£o, mais 
#forte Ã© a dependÃªncia de B em relaÃ§Ã£o a A. Uma convicÃ§Ã£o de 1 indica independÃªncia (ausÃªncia de associaÃ§Ã£o entre A e B).
```

```python
regras = association_rules(associacao, metric = 'lift')
regras.head()
```
