```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn.metrics import mean_absolute_error, mean_squared_error
import lightgbm as lgbm
```

```python
path = os.getcwd() + '\\Analise e Tratamento\\Dados\\'
df = pd.read_csv(path + 'housing.csv', sep =',' , encoding = 'iso-8859-1')
df.head(3)
```

```python
x = df.iloc[:, 0:3].values
y = df.iloc[:, 3].values
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.3, random_state = 0)
```

```python
x_treino.shape, y_teste.shape
```

```python
regressao_lgbm = lgbm.LGBMRegressor(num_leaves = 50, max_depth = 3, learning_rate = 0.1, n_estimators = 50)
regressao_lgbm.fit(x_treino, y_treino)
```

```python
#Coeficiente de Determinação dados de treino - r^2(Fala se os dados estão explicando o modelo)
regressao_lgbm.score(x_treino, y_treino)
```

```python
#Coeficiente de Determinação dados de teste - r^2
regressao_lgbm.score(x_teste, y_teste)
```

```python
previsoes = regressao_lgbm.predict(x_teste)
```

```python
valor_casa = regressao_lgbm.predict([[6,20,12]])
valor_casa
```

```python
#MÉTRICAS DE DESEMPENHO
```

```python
#Erro absoluto médio
abs(y_teste - previsoes).mean()
```

```python
mean_absolute_error(y_teste, previsoes)
```

```python
#Erro quadrático médio
mse = mean_squared_error(y_teste, previsoes)
mse
```

```python
#Raiz do erro quadrático médio
np.sqrt(mse)
```

```python
#VALIDAÇÃO CRUZADA
```

```python
kfold = KFold(n_splits = 15, shuffle = True, random_state = 5)
modelo = lgbm.LGBMRegressor(num_leaves = 50, max_depth = 3, learning_rate = 0.1, n_estimators = 50)
resultado = cross_val_score(modelo, x, y, cv = kfold)
print('Coeficiente de Determinação Médio - r^2: %.2f%%' % (resultado.mean() * 100))
```
