```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,2,100) # 100 valores igualmente espaçados entre 0 e 2
```

```python
y = 2 * x
```

```python
plt.plot(x,y,'-')
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.title('Titulo do Gráfico')

plt.show()
```

```python
y2 = x**2
```

```python
plt.plot(x,y2,'-')
plt.plot(x,y,'-')
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.title('Titulo do Gráfico')
plt.legend(['Parábola', 'Reta'])
plt.show()
```

```python
x2 = np.linspace(-1,1,100)
y3 = 2 * x2
y4 = x2**2
y5 = x2**3
y6 = np.sin(x2)
```

```python
fig, janela = plt.subplots(2,2) # duas linhas e duas colunas de janelas
janela[0,0].plot(x2,y3,'-')
janela[0,1].plot(x2,y4,'-')
janela[1,0].plot(x2,y5,'-')
janela[1,1].plot(x2,y6,'-')
fig.legend(['Reta', 'Parábola', 'Curva', 'Senóide'])
plt.show()
```
